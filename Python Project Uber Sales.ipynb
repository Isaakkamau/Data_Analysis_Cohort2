```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
file_path = 'C:/Users/USER/OneDrive/Documents/UBER.csv'
df = pd.read_csv(file_path)
df['Date']=pd.to_datetime(df['Date'])
plt.plot(df['Date'], df['Open'], color='red', label='Open')
plt.plot(df['Date'], df['Close'], color='yellow', label='Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Opening and Closing prices')
plt.xticks(rotation=75)
plt.legend()
plt.show()
```

```python
plt.plot(df['Date'], df['High'], color='brown', label='High')
plt.plot(df['Date'], df['Low'], color='yellow', label='Low')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('High and Low Prices')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.plot(df['Date'], df['Volume'], color='purple')
plt.xlabel('Date')
plt.ylabel('Trading Volume')
plt.title('Trend in Stock Trading Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
#we can conclude our analyses by saying that the market is neutral, the open and close prices for each day are on the same level
```
